```python
import os
import pandas as pd
```

```python
directory_path = r'<path_to_csv_directory>'
csv_name = r'<csv_name>'
target_path = r'<path_to_new_csv>'
desired_order = ['City', 'Link', 'Name', 'Rating', 'Price', 'Longitude','Latitude', 'Amenity']
```

```python
def extract_prices(price_str):
    try:
        price_int = int(price_str[price_str.index("₺")+1:].replace(".",""))
        return price_int
    except:
        return -1
```

```python
def extract_coordinates(coordinates_str):
    try:
        coordinates = coordinates_str.split(",")
        longitude = float(coordinates[0][2:-1])
        latitude = float(coordinates[1][2:-2])
        return longitude,latitude
    except:
        return -1.0,-1.0
```

```python
def extract_amenity(amenity_str):
    try:
        amenity_str = amenity_str.replace(", ",",").replace('\'',"").replace("[","").replace("]","")
        return amenity_str
    except:
        return None
```

```python
def extract_rating(rating_str):
    try:
        rating_str = float(rating_str.replace(",","."))
        return rating_str
    except:
        return -1.0
```

```python
df = pd.read_csv(directory_path + "/" + csv_name)
df.head()
```

```python
df = df.drop('Website', axis=1)
df = df.drop('Tel', axis=1)
df['Price'] = df['Price'].apply(extract_prices)
df.head()
```

```python
df['Longitude'], df['Latitude'] = zip(*df['Coordinates'].apply(extract_coordinates))
df = df.drop('Coordinates', axis=1)
df.head()
```

```python
df['Amenity'] = df['Amenity'].apply(extract_amenity)
df.head()
```

```python
df['Rating'] = df['Rating'].apply(extract_rating)
df.head()
```

```python
df = df[desired_order]
df.head()
```

```python
df.to_csv(target_path + "/" + csv_name[:csv_name.index('.')] + "_for_db.csv", index=False ,encoding='utf-8-sig')
```
