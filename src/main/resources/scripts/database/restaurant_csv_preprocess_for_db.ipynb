```python
import os
import pandas as pd
```

```python
directory_path = r'<path_to_csv_directory>'
csv_name = r'<csv_name>'
target_path = r'<path_to_new_csv>'
desired_order = ['City', 'Link', 'Name', 'Rating', 'Price_Lower', 'Price_Higher', 'Cuisine']
```

```python
def extract_prices(price_range_str):
    try:
        price_lower = int(price_range_str[1:price_range_str.index("-")-1].replace(".",""))
        price_higher = int(price_range_str[price_range_str.index("-")+3:].replace(".",""))
        return price_lower,price_higher
    except:
        return -1,-1
```

```python
df = pd.read_csv(directory_path + "/" + csv_name)
df.head()
```

```python
price_low, price_high = extract_prices(str(df['Price'][4]))
print(price_low)
print(price_high)
```

```python
df['Price_Lower'], df['Price_Higher'] = zip(*df['Price'].apply(extract_prices))
df.head()
```

```python
df = df.drop('Price', axis=1)
df.head()
```

```python
df = df[desired_order]
df.head()
```

```python
df.to_csv(target_path + "/" + csv_name[:csv_name.index('.')] + "_for_db.csv", index=False ,encoding='utf-8-sig')
```
