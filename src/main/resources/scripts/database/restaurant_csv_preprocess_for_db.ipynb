```python
def extract_prices(price_range_str):
    try:
        price_lower = int(price_range_str[1:price_range_str.index("-")-1].replace(".",""))
        price_higher = int(price_range_str[price_range_str.index("-")+3:].replace(".",""))
        return price_lower,price_higher
    except:
        return -1,-1
```

```python
def extract_coordinates(coordinates_str):
    try:
        coordinates = coordinates_str.split(",")
        longitude = float(coordinates[0])
        latitude = float(coordinates[1])
        return longitude,latitude
    except:
        return -1.0,-1.0
```

```python
def extract_rating(rating_str):
    try:
        rating = float(rating_str[rating_str.index(",")-1:rating_str.index(",")+1].replace(",","."))
        return rating
    except:
        return -1.0
```

```python
meal_types_to_move_cuisine = ['vegeterian', 'vegan', 'gluten_free']

for index, row in df.iterrows():
    for word in meal_types_to_move_cuisine:
        if word in str(row['meal_type']).lower():
            if pd.isna(row['cuisine']):
                df.at[index, 'cuisine'] = word
            else:
                df.at[index, 'cuisine'] += ', ' + word
            df.at[index, 'meal_type'] = str(row['meal_type']).replace(', ' + word, '').replace(word+',', '').replace(word, '').strip()

df['meal_type'] = df['meal_type'].apply(lambda x: '' if pd.isna(x) else x)

df.head()
```

```python
word = 'local_cusine'

for index, row in df.iterrows():
    if word in str(row['purpose']).lower():
        if pd.isna(row['cuisine']):
            df.at[index, 'cuisine'] = 'Yöresel mutfak'
        else:
            df.at[index, 'cuisine'] += ', ' + 'Yöresel mutfak'
        df.at[index, 'purpose'] = str(row['purpose']).replace(', ' + word, '').replace(word+',', '').replace(word, '').strip()

df['purpose'] = df['purpose'].apply(lambda x: '' if pd.isna(x) else x)
df.head()
```

```python
def change_cuisine_tokens(df, translation_map):
    def translate_cuisine(cuisine):
        if not isinstance(cuisine, str):
            return cuisine
        translated_cuisines = [translation_map.get(token.strip(), token.strip()) for token in cuisine.split(',')]
        return ', '.join(set(translated_cuisines))
    df['cuisine'] = df['cuisine'].apply(translate_cuisine)
    return df

df = change_cuisine_tokens(df, TurkishToEnlishMap)
```
