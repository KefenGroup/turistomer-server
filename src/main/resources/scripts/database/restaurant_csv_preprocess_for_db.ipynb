```python
import os
import pandas as pd
```

```python
directory_path = r'<csv_directory>'
csv_name = r'<csv_name>'
target_path = r'<path_to_new_csv>'
desired_order = ['city', 'link', 'name', 'rating', 'price_lower', 'price_higher', 'price_type','longitude', 'latitude', 'cuisine', 'meal_type','purpose']
```

```python
def extract_prices(price_range_str):
    try:
        price_lower = int(price_range_str[1:price_range_str.index("-")-1].replace(".",""))
        price_higher = int(price_range_str[price_range_str.index("-")+3:].replace(".",""))
        return price_lower,price_higher
    except:
        return -1,-1
```

```python
def extract_coordinates(coordinates_str):
    try:
        coordinates = coordinates_str.split(",")
        longitude = float(coordinates[0])
        latitude = float(coordinates[1])
        return longitude,latitude
    except:
        return -1.0,-1.0
```

```python
def extract_rating(rating_str):
    try:
        rating = float(rating_str[rating_str.index(",")-1:rating_str.index(",")+1].replace(",","."))
        return rating
    except:
        return -1.0
```

```python
df = pd.read_csv(directory_path + "/" + csv_name)
df.head()
```

```python
df = df.drop("Column1", axis=1 )
df = df.rename(columns={'cousine': 'cuisine'})
df = df.rename(columns={'mealType': 'meal_type'})
df = df.rename(columns={'priceType': 'price_type'})
df = df.rename(columns={'priceType': 'price_type'})

df['cuisine'] = df['cuisine'].astype(str).fillna('') + ", " + df['diets'].fillna('')
df = df.drop("diets", axis=1 )
df.head()
```

```python
df['price_lower'], df['price_higher'] = zip(*df['price'].apply(extract_prices))
df.head()
```

```python
df['longitude'], df['latitude'] = zip(*df['coordinates'].apply(extract_coordinates))
df.head()
```

```python
df['rating'] = df['rating'].apply(extract_rating)
df.head()
```

```python
df = df.drop('price', axis=1)
df = df.rename(columns={'cousine': 'cuisine'})
df = df[desired_order]
df.head()
```

```python
df['price_type'] = df['price_type'].fillna('')
df['meal_type'] = df['meal_type'].fillna('')
df['purpose'] = df['purpose'].fillna('')
```

```python
df.to_csv(target_path + "/" + csv_name[:csv_name.index('.')] + "_for_db.csv", index=False ,encoding='utf-8-sig')
```
